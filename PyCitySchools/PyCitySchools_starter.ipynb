```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
school_data_complete_filter = school_data_complete.loc[school_data_complete["school_name"]=="Huang High School"]
school_data_complete_filter
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# calculate total schools and budget
total_schools = len(school_data_complete["school_name"].unique())
total_budget = sum(school_data["budget"])

# calculate total students, different students may have same name
total_students = len(school_data_complete["student_name"])

# calculate avg scores
avg_math = sum(school_data_complete["math_score"]) / total_students
avg_reading = sum(school_data_complete["reading_score"]) / total_students
overall_passing = (avg_math + avg_reading) / 2

# calculate the passing percentages
math_passing_count = ([int(score) >= 70 for score in school_data_complete["math_score"]]).count(True)
reading_passing_count = ([int(score) >= 70 for score in school_data_complete["reading_score"]]).count(True)

# passing_count = ([int(score) >= 70 for score in school_data_complete["math_score"] & school_data_complete["reading_score"]]).count(True)
# passing_count
# passing_pct = passing_count / total_students

math_passing_pct = math_passing_count / total_students
reading_passing_pct = reading_passing_count / total_students

# create a dataframe to hold the results
scores_dicts = [{"Total Schools": total_schools, "Total Students": total_students, "Total Budget":total_budget, 
                 "Average Math Score":avg_math, "Average Reading Score":avg_reading, "% Passing Math":math_passing_pct,
                "% Passing Reading":reading_passing_pct, "% Overall Passing Rate":overall_passing}]
df_scores = pd.DataFrame(scores_dicts)
df_scores

# formatting the dataframe
df_scores = df_scores.style.format({
    'Total Students': '{:,}'.format,
    'Total Budget': '${:,}'.format,
    'Average Math Score':'{:.4}'.format,
    'Average Reading Score':'{:.4}'.format,
    '% Passing Math': '{:,.2%}'.format,
    '% Passing Reading': '{:,.2%}'.format
})
df_scores
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# create a groupby by school_name to calculate the data by schools
df_school_summary = school_data_complete.groupby(["school_name"])
school_type = df_school_summary['type'].unique()

# calculation for "Total Students"
total_students_school = df_school_summary["student_name"].count()

# calculation for "Total School Budget"
total_budget_school = df_school_summary["budget"].sum() / df_school_summary["budget"].count()

# calculation for "Per Student Budget"
per_student_budget = total_budget_school / total_students_school

# calculation for "Average Math Score" & "Average Reading Score"
avg_math_school = df_school_summary["math_score"].sum() / total_students_school
avg_reading_school = df_school_summary["reading_score"].sum() / total_students_school

# calculation for "% Passing Math", "% Passing Reading" & "Overall Passing Rate"
math_passing_count_school = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['math_score'].count()
reading_passing_count_school = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['reading_score'].count()

math_passing_pct_school = math_passing_count_school / total_students_school
reading_passing_pct_school = reading_passing_count_school / total_students_school
overall_passing_school = (math_passing_pct_school + reading_passing_pct_school)/2

# math_passing_count_school = ([int(i) >= 70 for i in df_school_summary["math_score"]]).count(True)
# reading_passing_count_school = ([int(i) >= 70 for i in df_school_summary["reading_score"]]).count(True)
# df_school_summary["math_score"].value_counts()

# display the results
school_dicts = {"School Type": school_type, "Total Students": total_students_school, "Total School Budget":total_budget_school, 
                 "Per Student Budget":per_student_budget, "Average Math Score":avg_math_school, "Average Reading Score":avg_reading_school, 
                 "% Passing Math":math_passing_pct_school, "% Passing Reading":reading_passing_pct_school, "% Overall Passing Rate":overall_passing_school}
df_schools = pd.DataFrame(school_dicts)

# sort the dataframe by decending order of "% Overall Passing Rate"
df_schools_sorted = df_schools.sort_values("% Overall Passing Rate",ascending=False)

# find out the top 5 performing schools
top_schools = df_schools_sorted.head(5)

# formatting the dataframe
top_schools = top_schools.style.format({
    'Total Students': '{:,}'.format,
    'Total School Budget': '${:,}'.format,
    'Per Student Budget': '${:,}'.format,
    'Average Math Score':'{:.4}'.format,
    'Average Reading Score':'{:.4}'.format,
    '% Passing Math': '{:,.2%}'.format,
    '% Passing Reading': '{:,.2%}'.format,
    '% Overall Passing Rate': '{:,.2%}'.format
})

top_schools

# # formate the column "School Type"
# top_schools["School Type"] = top_schools["School Type"].apply(lambda u: u[0])
# top_schools
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Set up bins
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

# apply the bins to previous dataframe for easier calculation
df_schools_sorted["Budget Range"] = pd.cut(df_schools_sorted["Per Student Budget"], spending_bins, labels=group_names)
df_gouped_budget = df_schools_sorted.groupby(["Budget Range"])


# re-calculate "Average Math Score" & "Average Reading Score" for "Budget Range"
avg_math_budget = df_gouped_budget["Average Math Score"].sum() / df_gouped_budget["Average Math Score"].count()
avg_reading_budget = df_gouped_budget["Average Reading Score"].sum() / df_gouped_budget["Average Reading Score"].count()

# re-calculate "% Passing Math", "% Passing Reading" & "Overall Passing Rate" for "Budget Range"
math_passing_pct_budget = df_gouped_budget["% Passing Math"].sum() / df_gouped_budget["% Passing Math"].count()
reading_passing_pct_budget = df_gouped_budget["% Passing Reading"].sum() / df_gouped_budget["% Passing Reading"].count()
overall_passing_budget = (math_passing_pct_budget + reading_passing_pct_budget)/2

# display the results
budget_dicts = {"Average Math Score":avg_math_budget, "Average Reading Score":avg_reading_budget, 
                 "% Passing Math":math_passing_pct_budget, "% Passing Reading":reading_passing_pct_budget, 
                "% Overall Passing Rate":overall_passing_budget}
df_budget = pd.DataFrame(budget_dicts)
df_budget
```
